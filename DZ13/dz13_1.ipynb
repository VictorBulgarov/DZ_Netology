## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
power_pribaltika = power[((power['country'] == 'Estonia') | (power['country'] == 'Latvia') | (power['country'] == 'Lithuansa')) 
      & ((power['year'] >= 2005) & (power['year'] <= 2010)) 
      & ((power['category'] == 21) | (power['category'] == 12) | (power['category'] == 4)) 
      & (power['quantity'] > 0)]
power_pribaltika.head(30)
```

```python
power_pribaltika['quantity'].sum()
```
