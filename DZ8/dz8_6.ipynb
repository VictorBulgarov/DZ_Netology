Будем парсить страницу со свежеми новостям на habr.com/ru/all/.
Вам необходимо собирать только те статьи, в которых встречается хотя бы одно требуемое ключевое слово. Эти слова определяем в начале кода в переменной, например:
KEYWORDS = ['python', 'парсинг']
Поиск вести по всей доступной preview-информации (это информация, доступная непосредственно с текущей страницы).
В итоге должен формироваться датафрейм вида: <дата> - <заголовок> - <ссылка>
Дополнительная часть (необязательная)
Улучшить скрипт так, чтобы он анализировал не только preview-информацию статьи, но и весь текст статьи целиком.
Для этого потребуется получать страницы статей и искать по тексту внутри этой страницы.
Итоговый датафрейм формировать со столбцами: <дата> - <заголовок> - <ссылка> - <текст_статьи>


```python
import requests
```

```python

import time
time.sleep(0.2)
```

```python

from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python

URL = 'http://habr.com/ru/all/'
KEYWORDS = ['python', 'парсинг', 'IT', 'облако']
```

```python
res = requests.get(URL)
```

```python
res.text
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
news_blocks = soup.find_all('article', class_='post post_preview')
```

```python
print(news_blocks)
```

```python

hub_news = pd.DataFrame()

for news in news_blocks:
    news_preview = news.find('ul', class_='post__hubs inline-list')
    news_preview_text = news_preview.text.lower()
    
    for req in KEYWORDS:
        if req in news_preview_text:
            data = news.find('span', class_='post__time').text
            title = news.find('h2', class_='post__title').text.strip('\n')
            link = news.find('h2', class_='post__title')
            links = link.find('a').get('href')
            row = {'Дата':data, 'Заголовок':title, 'Ссылка':links}
            hub_news = pd.concat([hub_news, pd.DataFrame([row])])
        else:
            continue
```

## Задание 2.
Обязательная часть
Написать скрипт, который будет проверять список e-mail адресов на утечку при помощи сервиса Avast Hack Ckeck. Список email-ов задаем переменной в начале кода:
EMAIL = [xxx@x.ru, yyy@y.com]
В итоге должен формироваться датафрейм со столбцами: <дата утечки> - <источник утечки> - <описание утечки>
Подсказка: сервис работает при помощи "скрытого" API. Внимательно изучите post-запросы.

```python
import json
```

```python

URL = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'
EMAIL = ['aaa@mail.ru', 'bbb@mail.com']
js_email = json.dumps({'emailAddresses': EMAIL})
```

```python
HEADERS = {
    'Vaar-Header-App-Build-Version': '1.0.0',
'Vaar-Header-App-Product': 'hackcheck-web-avast',
'Vaar-Header-App-Product-Name': 'hackcheck-web-avast',
'Vaar-Version': '0'
}
```

```python
req = requests.post(URL, data = js_email, headers = HEADERS)
```

```python
req.text
```

```python
dict_emails = req.json()
```

```python
print(dict_emails)
```

```python
df_mails = pd.DataFrame(columns = ('дата утечки', 'источник утечки', 'описание утечки'))

for mail in dict_emails['breaches'].values():
    df_mails = df_mails.append({'дата утечки': mail['publishDate'],
                                      'источник утечки': mail['site'],
                                      'описание утечки': mail['description']
                                     }, ignore_index = True)
    
df_mails
```
