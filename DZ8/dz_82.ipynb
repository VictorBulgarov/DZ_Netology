```python

def get_all_links(url, query, pages):
    URL = 'https://habr.com/ru/search'
    all_refs = []
    params = {'q': 'python парсинг'}
    for i in range(pages):
        params['page'] = i + 1
        res = requests.get(URL, params)
        time.sleep(0.3)
        soup = BeautifulSoup(res.text, 'html.parser')
        news_blocks = soup.find_all('a', class_='post__title_link')
        a_list = list(map(lambda x: x.get('href'), news_blocks))
        all_refs += list(map(lambda x: x, a_list))

#     print (a_list)
    return all_refs

all_links = get_all_links(URL, 'python парсинг', 2)
print(all_links)
print(len(all_links))
```

```python
import pandas as pd
```

```python
for link in all_links:
    soup = BeautifulSoup(requests.get(link).text, 'html.parser')
    time.sleep(0.3)
    date = pd.to_datetime(soup.find('span', class_='post__time').get('data-time_published'), dayfirst=True).date()
    print(date)
    title = soup.find('h1', class_='post__title').text
    print(title)
    text = soup.find('div', class_='post__text').text
    print(text)
```

```python

hub_news = pd.DataFrame()
for link in all_links:
    soup = BeautifulSoup(requests.get(link).text, 'html.parser')
    time.sleep(0.3)
    date = pd.to_datetime(soup.find('span', class_='post__time').get('data-time_published'), dayfirst=True).date()
    title = soup.find('h1', class_='post__title').text
    text = soup.find('div', class_='post__text').text
    links = link
    row = {'date': date, 'title': title, 'text': text, 'links': links}
    hub_news = pd.concat([hub_news, pd.DataFrame([row])])
hub_news  
```
