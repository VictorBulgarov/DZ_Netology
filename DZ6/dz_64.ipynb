Исходные данные имеют следующую структуру:

documents = [

directories = {
‘1’: [‘2207 876234’, ‘11-2’],
‘2’: [‘10006’],
‘3’: []
}

Общие требования к программе:

```python
documents = [
        {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
        {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
        {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
      ]
```

```python
directories = {
        '1': ['2207 876234', '11-2'],
        '2': ['10006'],
        '3': []
      }
```

```python
def get_name():
    
    """Функция get_name определяет полку по номеру документа"""
    
    number = input('Введите номер документа ')
    for i in documents:
        if i.get("number") == number:
            print((f"Владелец документа {i.get('name')}"))
            return 
    print(f'Документ не найден в базе')
```

```python
def get_shelf():
    """Функция get_shelf определяет владельца документа по номеру"""

    number = input('Введите номер документа ')
    for i in directories:
        if number in directories.get(i):
            print(f"Документ хранится на полке {i}")
            return
    else:
        print('В полках документа с данным номером нет.')
```

```python
def get_info():
    
    """Функция get_info выводит полную информацию по документам"""
    
    for shelf in directories:
        for number in directories[shelf]:
            found = next(i for i in documents if i['number']==number)
            print(f'№: {number}, тип:{found["type"]}, владелец:{found["name"]}, полка хранения: {shelf}')
```

```python
def add_shelf():
    
    """Функция add_shelf добавляет новую полку"""
    
    shelf = input("Введите номер полки: ")
    for i in directories.items():
        if shelf in i[0]:
            print(f"Такая полка уже существует. Текущий перечень полок: {list(directories.keys())}")
            break
    else:
        directories[shelf] = []
        print(f"Полка добавлена. Текущий перечень полок: {list(directories.keys())}")
```

```python
def del_shelf(): 
    
    """Функция del_shelf удаляет полку"""
    
    number = input("Введите номер полки: ")
    shelf = list(directories.values())
    if number not in directories:
        return "Полки не существует"
    elif directories[number]:
        print (f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {list(directories.keys())}")
    else: 
        del(directories[number])
        print ("Полка удалена. Текущий перечень полок:", list(directories.keys()))
    return 
```

```python
def main():
    
    """Функция main общая функция для всей программы"""
    
    
    while True:
        print('Возможные команды: p, s, l, as, ds, q')
        command = input('Введите команду : ')
 
        if command == 'p':
            get_name()
    
        elif command == 's':
            get_shelf()
    
        elif command == 'l':
            get_info()
    
        elif command == 'as':
            add_shelf()
            
        elif command == 'ds':
            del_shelf()
    
        elif command == "q":
            break
```

```python
main()
```
